```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Team1'] = le.fit_transform(df['Team1'])
df['Team2'] = le.fit_transform(df['Team2'])
df['Winner'] = le.fit_transform(df['Winner'])
df['Toss_Winner'] = le.fit_transform(df['Toss_Winner'])
df['Toss_Decision'] = le.fit_transform(df['Toss_Decision'])
df['City'] = le.fit_transform(df['City'])
```

```python
X = df[['Team1','Team2','Toss_Winner','Toss_Decision','City','Season']]
```

```python
y_class = df['Winner']
y_reg = df['Win_Margin']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.2, random_state=42)
_, _, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train_c)
```

```python
y_pred_c = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_score(y_test_c, y_pred_c)
```

```python
confusion_matrix(y_test_c, y_pred_c)
```

```python
classification_report(y_test_c, y_pred_c)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train_r)
```

```python
y_pred_r = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mean_absolute_error(y_test_r, y_pred_r)
```

```python
mean_squared_error(y_test_r, y_pred_r)
```

```python
r2_score(y_test_r, y_pred_r)
```
